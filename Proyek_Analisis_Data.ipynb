### Cleaning Data

Setelah melakukan analisis data pada data assesing sebelumnya, setiap dataset mempunyai bagian - bagian yang kosong di tiap colom : PM2.5, PM10, SO2, NO2, CO, O3. TEMP, PRES, DEWP, RAIN, wd, dan WSPM. Pada bagian cleaning data ini saya akan melakukan pembersihan data seperti menghapus data duplikat dan mengisi data yang kosong.

```python
all_df = pd.concat([
    aotizhongxin_df.assign(station='Aotizhongxin'),
    changping_df.assign(station='Changping'),
    dingling_df.assign(station='Dingling'),
    dongsi_df.assign(station='Dongsi'),
    guanyuan_df.assign(station='Guanyuan'),
    gucheng_df.assign(station='Gucheng'),
    huairou_df.assign(station='Huairou'),
    nongzhanguan_df.assign(station='Nongzhanguan'),
    shunyi_df.assign(station='Shunyi'),
    tiantan_df.assign(station='Tiantan'),
    wanliu_df.assign(station='Wanliu'),
    wanshouxigong_df.assign(station='Wanshouxigong')
], ignore_index=True)
```

Menggabungkan seluruh dataset menjadi satu agar memudahkan pembersihan data dan tidak memakan banyak waktu.

```python
all_df['datetime'] = pd.to_datetime(all_df[['year', 'month', 'day', 'hour']])
```

Untuk memudahkan dalam pemrosesan data yang berhubungan dengan waktu nantinya, saya membuat kolom baru bernama datetime yang dihasilkan dari kolom year, month, day, dan hour.

```python
all_df.drop_duplicates(inplace=True)
```

Menghilangkan seluruh data duplikat

```python
all_df.duplicated().sum()
```

```python
all_df.dtypes
```

```python
all_df['wd'] = all_df['wd'].astype(str)
```

Dikarenakan adanya kesalahan tipe data pada kolom 'wd' maka saya akan mengubahnya terlebih dahulu menjadi '`object`'

```python
# Mengganti missing values pada kolom 'PM2.5' dengan rata-rata kolom 'PM2.5'
all_df['PM2.5'].fillna(all_df['PM2.5'].mean(), inplace=True)

# Mengganti missing values pada kolom 'PM10' dengan rata-rata kolom 'PM10'
all_df['PM10'].fillna(all_df['PM10'].mean(), inplace=True)

# Mengganti missing values pada kolom 'SO2' dengan rata-rata kolom 'SO2'
all_df['SO2'].fillna(all_df['SO2'].mean(), inplace=True)

# Mengganti missing values pada kolom 'NO2' dengan rata-rata kolom 'NO2'
all_df['NO2'].fillna(all_df['NO2'].mean(), inplace=True)

# Mengganti missing values pada kolom 'CO' dengan rata-rata kolom 'CO'
all_df['CO'].fillna(all_df['CO'].mean(), inplace=True)

# Mengganti missing values pada kolom 'O3' dengan rata-rata kolom 'O3'
all_df['O3'].fillna(all_df['O3'].mean(), inplace=True)

# Mengganti missing values pada kolom 'TEMP' dengan rata-rata kolom 'TEMP'
all_df['TEMP'].fillna(all_df['TEMP'].mean(), inplace=True)

# Mengganti missing values pada kolom 'PRES' dengan rata-rata kolom 'PRES'
all_df['PRES'].fillna(all_df['PRES'].mean(), inplace=True)

# Mengganti missing values pada kolom 'DEWP' dengan rata-rata kolom 'DEWP'
all_df['DEWP'].fillna(all_df['DEWP'].mean(), inplace=True)

# Mengganti missing values pada kolom 'RAIN' dengan rata-rata kolom 'RAIN'
all_df['RAIN'].fillna(all_df['RAIN'].mean(), inplace=True)

# Mengganti missing values pada kolom 'wd' dengan rata-rata kolom 'wd'
all_df['wd'].fillna(all_df['wd'].mode(), inplace=True)

# Mengganti missing values pada kolom 'WSPM' dengan rata-rata kolom 'WSPM'
all_df['WSPM'].fillna(all_df['WSPM'].mean(), inplace=True)
```

Saya menggunakan metode imputation untuk mengisi value data yang hilang, pada kolom yang bertipe `int` dan `float` menggunakan `mean()` dan pada kolom bertipe `object` menggunakan `mode()`

```python
all_df.isnull().sum()
```

```python
all_df.describe()
```

**Insight:**
- Setelah melakukan imputasi,` all_df.isnull().sum()` akan menghasilkan 0 untuk semua kolom. Ini menunjukkan semua missing values sudah terisi dan dataset siap untuk dianalisis lebih lanjut.

## Exploratory Data Analysis (EDA)

Pertanyaan bisnis yang akan dijawab:

1. Seberapa besar perubahan polutan setiap tahunnya pada tiap stasiun?

2. Stasiun mana yang memiliki rata-rata PM2.5 tertinggi pada setiap tahun?

3. Apakah terdapat hubungan antara tingkat PM2.5 dengan rata-rata tingkat polutan lainnya?






Untuk mempermudah menjawab pertanyaan tentang tren rata-rata polutan di setiap stasiun selama 5 tahun, saya membuat pivot table bernama `pollution_trend_pivot` menggunakan fungsi `pivot_table()` dari library Pandas.

Pivot table ini dibuat dengan menggunakan DataFrame `all_df` yang berisi data kualitas udara dari semua stasiun.

```python
pollution_trend_pivot = all_df.pivot_table(
    values=['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3'],
    index=['station', 'year'],
    aggfunc='mean'
)

pollution_trend_pivot['total_pollution'] = pollution_trend_pivot.sum(axis=1)
pollution_trend_pivot = pollution_trend_pivot.reset_index()

pollution_trend_pivot
```

Berdasarkan pivot table yang dihasilkan, dapat terlihat bahwa tren rata-rata polutan di setiap stasiun pengukuran selama periode 5 tahun bervariasi. Rata - rata pada setiap stasiun mengalami kenaikan dari 2013 hingga ke 2017. Perbedaan tren ini dapat disebabkan oleh berbagai faktor, seperti lokasi stasiun, aktivitas industri di sekitar stasiun, dan kebijakan pengendalian polusi udara yang diterapkan.

Untuk mempermudah menjawab pertanyaan tentang stasiun mana yang memiliki rata-rata PM2.5 tertinggi pada setiap tahun, saya membuat pivot table bernama `pm25_pivot` menggunakan fungsi `pivot_table()` dari library Pandas.

Nilai rata-rata PM2.5 di setiap stasiun pada setiap tahun dihitung menggunakan fungsi `pivot_table()` dan kemudian diurutkan dari yang tertinggi ke terendah menggunakan fungsi `sort_values()`.

```python
pm25_pivot = all_df.pivot_table(
    values='PM2.5',
    index='year',
    columns='station',
    aggfunc='mean'
).sort_values(by=[2013, 2014, 2015, 2016,2017], ascending=False, axis=1)


pm25_pivot
```

Dapat terlihat bahwa stasiun - stasiun yang memiliki PM2.5 tertinggi berada pada bagian kiri dan menurun ke bagian kanan. Wanliu, Dongsi, dan Nongzhanguan menempati daftar posisi dengan PM2.5 tertinggi.

Untuk melihat hubungan antara PM2.5 dengan rata-rata polutan lainnya, saya membuat pivot table bernama `correlation_pivot` menggunakan fungsi `corr()` dari library Pandas. Pivot table ini menampilkan matriks korelasi antara PM2.5 dan polutan lainnya (PM10, SO2, NO2, CO, dan O3) yang dihitung menggunakan fungsi `corr(`) pada DataFrame `all_df` yang hanya berisi kolom-kolom polutan yang ingin dianalisis.

```python
correlation_pivot = all_df[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']].corr()

correlation_pivot
```

Dari pivot table tersebut menunjukkan PM2.5 memiliki korelasi positif yang kuat dengan PM10, CO, NO2, dan SO2. Ini menunjukkan bahwa polutan-polutan ini cenderung muncul bersamaan di udara. PM2.5 memiliki korelasi negatif yang lemah dengan O3, yang menunjukkan sedikit hubungan terbalik antara kedua polutan tersebut.

**Insight**
*   PM2.5 memiliki korelasi positif yang kuat dengan PM10, CO, dan NO2, mengindikasikan bahwa polutan-polutan ini cenderung muncul bersamaan.
*   Terdapat perbedaan yang signifikan dalam tingkat rata-rata PM2.5 di antara stasiun-stasiun pengukuran, menunjukkan variasi kualitas udara di berbagai wilayah di Beijing.



## Visualization & Explanatory Analysis

### Pertanyaan 1: Seberapa besar perubahan polutan setiap tahunnya pada tiap stasiun?

Disini saya membagi tiap staisun menjadi satu chart agar lebih mudah dipahami. Untuk memudahkan pembuatan, saya menggunakan perulangan dan memasukkan hasil pivot table yang telah dibuat sebelumnya yaitu `pollution_trend_pivot` kedalam variabel `stations`.






```python
stations = pollution_trend_pivot['station'].unique()

for station in stations:
    station_data = pollution_trend_pivot[pollution_trend_pivot['station'] == station]
    plt.figure(figsize=(10, 6))
    sns.lineplot(x='year', y='total_pollution', data=station_data)
    plt.title(f'Tren Perubahan Polutan di Stasiun {station}')
    plt.xlabel('Tahun')
    plt.ylabel('Jumlah Rata-rata Polutan')
    plt.xticks(range(2013, 2018))
    plt.show()
```

### Pertanyaan 2: Stasiun mana yang memiliki rata-rata PM2.5 tertinggi pada setiap tahun?

Disini untuk menjawab pertanyaan nomor 2, saya menggunakan parameter `plt.bar()` untuk membuat bar chart dengan `stations` sebagai label sumbu x dan nilai PM2.5 dari `pm25_pivot` sebagai tinggi bar.


```python
years = pm25_pivot.index
stations = pm25_pivot.columns


for year in years:
    plt.figure(figsize=(10, 6))
    plt.bar(stations, pm25_pivot.loc[year].sort_values(ascending=False))
    plt.title(f'Rata-rata PM2.5 Tertinggi di Setiap Stasiun - Tahun {year}')
    plt.xlabel('Stasiun')
    plt.ylabel('Rata-rata PM2.5')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

### Pertanyaan 3: Apakah terdapat hubungan antara tingkat PM2.5 dengan rata-rata tingkat polutan lainnya?


Disini saya menggunakan fitur `heatmap()` yang disediakan oleh library seaborn agar memudahkan dalam mengubah pivot table correlation menjadi bentuk visual data

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_pivot, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasi antara PM2.5 dan Polutan Lainnya')
plt.show()
```

PM2.5 memiliki korelasi positif yang kuat dengan PM10, CO, dan NO2, yang artinya jika konsentrasi PM2.5 tinggi, maka konsentrasi polutan-polutan tersebut juga cenderung tinggi. Korelasi positif sedang terlihat dengan SO2, mengindikasikan hubungan yang cukup erat, namun tidak sekuat polutan sebelumnya. Korelasi dengan O3 cenderung negatif dan lemah, menunjukkan bahwa peningkatan PM2.5 mungkin sedikit berhubungan dengan penurunan O3, atau hubungannya tidak terlalu signifikan.

**Insight**

*    Visualisasi menunjukkan tren peningkatan polusi udara di sebagian besar stasiun selama periode 2013-2017.
*   Stasiun-stasiun seperti Wanliu, Dongsi, dan Nongzhanguan secara konsisten memiliki rata-rata PM2.5 tertinggi selama periode pengamatan.
